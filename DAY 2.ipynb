```python
#calculate the number of upper and lower case alphabets in a given string
#string:She Sells Seashells by the Seashore
def calc_string(x):
  lower=0
  upper=0
  for i in x:
    if i.isupper()==True:
      upper+=1
    else:
      lower+=1
    return lower,upper
  print(calc_string("She Sells Seashells by the Seashore"))
```
