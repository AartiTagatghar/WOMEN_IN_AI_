```python
l=[1,3,"AARTI",5.5,5.6,"NAME"]
intlist=[]
floatlist=[]
strlist=[]

for i in l:
  if type(i)==int:
    intlist.append(i)
  elif type(i)==float:
    floatlist.append(i)
  elif type(i)==str:
    strlist.append(i)
print(intlist, floatlist, strlist)
```

```python
def recursive_fibonacci(n):
    if n <= 1:
        return n
    else:
        return(recursive_fibonacci(n-1) + recursive_fibonacci(n-2))
n_terms = 5
if n_terms <= 0:
    print("Invalid input ! Please input a positive value")
else:
    print("Fibonacci series:")
for i in range(n_terms):
    print(recursive_fibonacci(i))
```

```python
def recur_fact(n):
  if n==1:
    return(n)
  else:
    return n*recur_fact(n-1)
n=5
print(recur_fact(n))
```
